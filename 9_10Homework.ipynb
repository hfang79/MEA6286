```python
import numpy as np 
from matplotlib import pyplot
T=100.0
dt=0.05
t=np.arange(0.0,T+0.05,dt)
z0=100.0
b0=10.0
zt=100.0
g=9.81
u = np.array([z0, b0])
N = int(T / dt) + 1

z=np.zeros(N)
z[0]=z0
a=1
while a < N:
    rhs = np.array([u[1], g * (1 - u[0] / zt)])
    u = u + dt * rhs
    z[a] = u[0]
    a +=1
```
